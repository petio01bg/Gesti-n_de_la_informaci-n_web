```python
import requests

#Fuentes de informacion de categorias para opcion 2
def categorias():
    url = "https://vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com/api/vaccines/get-everything"

    headers = {
        "X-RapidAPI-Key": "6132525ee2msh4a84008e73591ddp1e43eejsn1c139e918c0b",
        "X-RapidAPI-Host": "vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com"
    }

    response = requests.request("GET", url, headers=headers)
    return response

#Fuentes de informacion de tratamientos para opcion 2
def tratamientos():
    url = "https://vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com/api/vaccines/get-all-treatment"

    headers = {
        "X-RapidAPI-Key": "6132525ee2msh4a84008e73591ddp1e43eejsn1c139e918c0b",
        "X-RapidAPI-Host": "vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com"
    }

    response = requests.request("GET", url, headers=headers)
    return response

#Fuentes de informacion de vacunas para opcion 2 y 4
def vacunas():
    url = "https://vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com/api/vaccines/get-all-vaccines"

    headers = {
        "X-RapidAPI-Key": "6132525ee2msh4a84008e73591ddp1e43eejsn1c139e918c0b",
        "X-RapidAPI-Host": "vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com"
    }

    response = requests.request("GET", url, headers=headers)
    return response

#Mostrar informacion del opcion 2
def info(i):
    print("developerResearcher:",i["developerResearcher"])
    print("trimedName: ",i["trimedName"])
    print("category:",i["category"])
    print("trimedCategory:",i["trimedCategory"])
    print("phase:",i["phase"])
    print("nextSteps:",i["nextSteps"])
    print("description:",i["description"])
    print("funder:",i["funder"])

#Mostrar el contenido de la lista    
def mostrarLista(lista):
    j = 1
    for i in lista:
        print(str(j) + ". " + i)
        j += 1

#Eliminar las etiquetas de un texto
def eliminarEtiquetas(texto):
    pila = list()
    result = ""
    for i in texto:
        if(i == "<"):
            pila.append(i)
        elif(i == ">"):
            pila.pop()
        elif(len(pila) == 0):
            result += i
    return result

#Mostrar informacion de la vacuna
def vacunaInfo(i):
    print("\n")
    for key in i:
        print(key,":", i[key])

#Preguntar si quiere realizar otra busqueda
def otraBusqueda():
    print("\nQuiere volver a realizar otra busqueda(SI/NO):")
    respuesta = input()
    print("\n")
    if(respuesta.upper()=="NO"):
        return False
    return True
```

```python
import requests
import json

def opcion1():
    url = "https://vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com/api/npm-covid-data/"

    headers = {
        "X-RapidAPI-Key": "6132525ee2msh4a84008e73591ddp1e43eejsn1c139e918c0b",
        "X-RapidAPI-Host": "vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com"
    }

    response = requests.request("GET", url, headers=headers)

    con = response.json()
    ok=True

    while(ok):

        li = list()
        paises =list()
        
        #Procesar los continentes
        for i in con:
            if(i["Continent"] not in li and i["Continent"] != "" and i["Continent"] != "All"):
                li.append(i["Continent"])
        
        #Mostrar los continentes y seleccionar uno
        print("\nSeleccione un continente:")
        mostrarLista(li)
        continente = input()
        print("\n")
        
        #Procesar los paises del dicho continente
        for i in con:
            if(continente==i["Continent"]):
                paises.append(i["Country"])

        #Mostrar paises y seleccionar uno
        print("Seleccione un pais:")
        mostrarLista(paises)
        pais = input()
        print("\n")

        #Buscar el pais y mostrar su informacion
        for i in con:
            if(i["Continent"]==continente and i["Country"] == pais):
                print("total_cases: " + str(i["TotalCases"]))
                print("new_cases: " + str(i["NewCases"]))
                print("total_deaths: " + str(i["TotalDeaths"]))
                print("new_deaths: " + str(i["NewDeaths"]))
                print("total_tests: " + str(i["TotalTests"]))
                break

        #Preguntar por realizar otra busqueda
        ok = otraBusqueda()
```

```python
import requests
import json

def opcion2():
    cat = categorias().json()
    treat = tratamientos().json()
    vac = vacunas().json()

    ok=True

    while(ok):
        
        li = list()
        
        #Seleccionar vacuna o tratamiento
        print("\nSelecciona:")
        print("1. Vaccine")
        print("2. Treatment")
        opcion=input()
        print("\n")
        
        #Procesar las categorias
        for i in cat:
            if(opcion == i["treatmentVsVaccine"]):
                if(i["trimedCategory"] not in li):
                    li.append(i["trimedCategory"])
        
        #Mostrar las categorias y seleccionar uno
        print("\nSeleccione la categoria:\n")
        mostrarLista(li)
        cate=input()
        
        #Seleccionar laboratorio en funcion de la opcion
        print("\nSeleccione el laboratorio:\n")
        j=1
        
        #Si es vacuna
        if(opcion == "Vaccine"):
            for i in vac:
                #Si la "cate" coincide con trimedCategory
                if(cate == i["trimedCategory"]):
                    #Mostrar el developerResearcher
                    print(str(j) + ". " + i["developerResearcher"])
                    j+=1
                    
            #Seleccionar el developerResearcher
            nombre = input()
            print("\n")
            for i in vac:
                #Si el "nombre" coincide con developerResearcher
                if(nombre == i["developerResearcher"]):
                    #Mostrar informacion del laboratorio
                    info(i)
                    break
                    
        #Si es tratamiento
        else:
            for i in treat:
                #Si la "cate" coincide con trimedCategory
                if(cate == i["trimedCategory"]):
                    #Mostrar el developerResearcher
                    print(str(j) + ". " + i["developerResearcher"])
                    j+=1
                    
            #Seleccionar el developerResearcher
            nombre = input()
            print("\n")
            for i in treat:
                #Si el "nombre" coincide con developerResearcher
                if(nombre == i["developerResearcher"]):
                    info(i)
                    break

        #Preguntar por realizar otra busqueda            
        ok = otraBusqueda()
```

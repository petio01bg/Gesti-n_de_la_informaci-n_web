A continuación, debeis calcular la distancia entre cada monumento y el lugar indicado. Esto se puede hacer usando la siguiente función que os propociono. 

En el dataset usado, se encuentra la longitud y latitud de cada monumento.

Para leer los archivos usa un código como el siguiente:

```python
#Busca si un elemento pertenece a la lista
def comprobar(lista, buscado):
    for i in lista:
        if(buscado == i):
            return True
    return False

#Muestra el contenido de la lista
def mostrarLista(lista):
    for i in lista:
        print(i)
    print("\n")

#Muestra la informacion del Monumento
def infoMonumento(lista):
    for i in lista:
        print("Nombre: ", i[0])
        print("Subtipo: ", i[1])
        print("Descripción: ", i[2])
        print("Fecha: ", i[3])
        print("Autores: ", i[4])
        print("Nombre_via: ", i[5])
        print("\n")

#Busca en los nombres y en los descripciones si contiene alguna de las palabras claves
def contiene(nombre, descripcion, palabraC):
    for i in palabraC:
        if((nombre.find(i) > -1) or (descripcion.find(i) > -1)):
            return True
    return False   

#Comprueba si "palabra" es vacia
def vacio(palabra):
    return palabra == ""
    
from math import radians, cos, sin, asin, sqrt
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    # Radius of earth in kilometers is 6371
    km = 6371* c
    return km
```
